# Demonstration - simulering av jonbytare i BPL

Följande reaktioner i mobilfas är den dynamiska modellen baserad på:

$$
\eqalign{
\begin{cases}
P + ES \rightleftharpoons PS + E\cr
A + ES \rightleftharpoons AS + E
\end{cases}}
$$

Här står $P$ för aktuelt produkt protein i mobilfas och $ES$ är bindning "site" och $PS$ är bundet protein och $E$ är lösta joner i mobilas. På samma sätt står $A$ för antagonist protein och följer liknande rekationer. Vi förutsätter att $pH$ är konstant.

Reaktionshastigheterna är $k_1$, $k_2$ för proteinet $P$ och $k_3$, $k_4$ för $A$.
Den totala kapaciteten av bindnings site är $Q_{av}$.

Jämviktskoncentrationen är $[PS] = {k_1}/{k_2} \cdot [ES]/[E] \cdot [P] $ och på samma sätt för $[AS]$.


Det är därför naturligt att förändra parametriseringen av bindnings-reaktionera så att $k_P$ = $k_1/k_2$ och pss för $k_A$. Denna parameter beskriver balansen vid jämvikt och ett mått på bindningens "styrka". Den är ungefär proportionell mot differensen mellan proteinets $pI$ och resinets $pH$ värde.

Det är också naturligt att identifiera tidskontanten för inbindning $T_P$ = $1/k_1$ och denna parameter beskrvier alltså kinetiken. Kinetiken påverkas av en rad faktorer och ej så tydlig tolkning av vad som dominerar $T_P$ som bindningens styrka $K_P$.

Kolonnen approximeras med 8 stycken seriekopplade tankar med ett gemensamt flöde $VFR$. Valet av antal seriekopplade tankar är något godtyckligt och en fördubbling ger något spetsigare toppar.

Modellen tar inte hänsyn til kolonnens geometri utan representeras internt som en homogen volym för mobilfasen. Dock har vi en parametrisering i termer av kolonnen höjd, diameter och andel som är mobilfas, och utifrån detta beräknas mobilfas-volymen. I branchen är det vanligt att ange flödet i termer av linear flow rate $LFR$ och det har också införts.

Modellen förutsätter ett homogent flöde. I rapporten nedan diskuteras kortfattat möjligheterna att utöka modellen för ett inhomogent flöde med långsammare flöde nära kolonnväggen och det ger lite bredare toppare. Denna utvidgning av modellen är inte genomförd här

## 1 Körning med relevanta parametrar

```python
# Körning med rimliga parametrar för körning med bibehållna parametrar för gelen k1..k4 men Q_av ökad

# Beräkning av V och VFR utifrån givna h, d,  lfr, samt default andel mobilfas x_m = 0.30
from numpy import pi
h = 20.0
d = 1.261
a = pi*(d/2)**2
V = h*a
print('V =', np.round(V,1), '[mL]')
lfr = 48
VFR = a*lfr/60
print('VFR =', np.round(VFR,1), '[mL/min]')
```

```python
# Inkommande material konc
par(P_in = 1.0, A_in = 1.0)

# Resin parametrar
K_P = 6.0; T_P = 3.33
K_A = 0.167; T_A = 20.0

par(k1=1/T_P, k2 = 1/(K_P*T_P), k3 = 1/T_A, k4 = 1/(K_A*T_A), Q_av = 6.0)

# Kolonnens storlek
par(height = h, diameter = d)

# Salt konc buffert för desorption samt initial salt konc i kolonnnen
par(E_in_desorption_buffer = 8.0)
init(E_0 = 50)

# Linear flow rate
par(LFR = lfr)

# Omslagspunkter räknat i kolonn-volymer V
par(scale_volume=True)
par(start_adsorption = 1.0*V, stop_adsorption = 1.5*V)
par(start_desorption = 2.5*V, stationary_desorption = 5.5*V, stop_desorption = 7.5*V)
par(start_pooling = 3.7*V, stop_pooling = 7.5*V)

# UV-styrning avslagen
par(start_uv=-1, stop_uv=-2)

# Plot och simulering
newplot(title='Hela cykeln: ekvilibrering, tvätt, desorption, poolning', \
        plotType='Elution-conductivity-vs-CV-combined-all')

simu(8.0*V/VFR)
```

**Kommentar** De fem nedersta kurvorna visar uppmätt salt-koncentration samt flödena från sampel och buffert1, därefeter salt-gradeint med buffert2, och nederst material som sköras genom poolning.

## 2 Variera K_A  - speglar bindningens styrka för A

```python
# Simulering och plot
newplot(title='Inverkan av variation av kA=k3/k4 - inbidningens styrka',\
        plotType='Elution-conductivity-vs-CV-combined-all')

T_A=20.0

K_A=0.001; par(k3=1/T_A, k4=1/(T_A*K_A)); simu(8.0*V/VFR)  # k3/k4=1.0
K_A=0.1; par(k3=1/T_A, k4=1/(T_A*K_A)); simu(8.0*V/VFR)  # k3/k4
K_A=0.2; par(k3=1/T_A, k4=1/(T_A*K_A)); simu(8.0*V/VFR)  # k3/k4=0.2
K_A=1.0; par(k3=1/T_A, k4=1/(T_A*K_A)); simu(8.0*V/VFR)  # k3/k4=1.0

    
# Restore default value
par(k3=0.05, k4=0.3)
```

**Kommentar** Vi ser att då $k_A$ ökar så binds A allt hårdare i kolonnen under absorptions-faen och mindre material sköljs ut och den första toppen därför allt lägre. Under desoprtionfasen ser vi på motsvarande sätt större mängd A frigörs. Topparna ungefär lika stora men arean under toppen som svarar mot mängd material, ökar. 

## 3 Variera T_A och behåll K_A konstant - speglar kinetiken

```python
# Simulering och plot
newplot(title='Inverkan av variation k3 med k_A konstant',\
        plotType='Elution-conductivity-vs-CV-combined-all')

K_A = 0.167

T_A=20.0; par(k3=1/T_A, k4=1/(T_A*K_A)); simu(8.0*V/VFR)  # k3/k4=0.1
T_A=33.0; par(k3=1/T_A, k4=1/(T_A*K_A)); simu(8.0*V/VFR)  # k3/k4=0.2
T_A=100; par(k3=1/T_A, k4=1/(T_A*K_A)); simu(8.0*V/VFR)  # k3/k4=1.0
    
# Restore default value
par(k3=0.05, k4=0.3)
```

**Kommentar** Då k3 minskar med kA konstant ändras bara kinetiken och inbindning och frisläppning går långsammare. Mängden material som binds respektive frisläpps senare är densamma.

## 4 Variera pH_resin - efterliknar pH förändring kanske?

```python
# Definiera hur pH_resin påverkar K_P och K_A

def KP_pH_sensitivity(pH_resin):
    pI_P = 8.0
    K_P_nom = 0.0
    delta_pH = 6.0
    return K_P_nom + delta_pH*(pI_P-pH_resin)

def KA_pH_sensitivity(pH_resin):
    pI_A = 7.167
    K_A_nom = 0.0
    delta_pH = 1.0
    return K_A_nom + delta_pH*(pI_A-pH_resin)
```

```python
# Inverkan av pH_resin på K_A och T_A på förenklat sätt

# Vi default värden på kinetiken men skriver ut dem för tydlighetens skull
T_P = 3.33
T_A = 20.0

# Simulering och plot
newplot(title='Inverkan av K_P och K_A genom pH_resin - inbindningens styrka',\
        plotType='Elution-conductivity-vs-CV-combined-all')

for pH in [7.0, 6.8, 6.6]: 
    par(k2=1/(T_P*KP_pH_sensitivity(pH)), k4=1/(T_A*KA_pH_sensitivity(pH))); 
    simu(8.0*V/VFR)
    
# Restore default value
par(k1=0.3, k2= 0.05, k3=0.05, k4=0.3)
```

**Kommentar** Vi ser att då pH_resin minskar så binds antagonisten in mer och upplösningen mellan topparna också något sämre. 

Detta är en högst förenklad bild av hur pH_resin eventuellt skulle kunna påverka båda topparna och vore intressant att fånga upp lite bättre hur det fungear i verkligheten. En rimlig formulering av pH-berodendet skulle kunna bidrat till att senare skapa ett mer relevant optimeringsproblem.

## 5 Variera kapacitet Q_av - genombrottskurva 

```python
# Experiment för att kolla upp Q_av
par(P_in=1, A_in=0)
init(E_0 = 0)
par(Q_av=6.0)

par(scale_volume=True, start_adsorption=1*V, stop_adsorption=4.0*V)
par(start_desorption=10*V, stationary_desorption=10.5*V, stop_desorption=11*V)
par(start_pooling=11*V, stop_pooling=12*V)

newplot(title='Inverkan av förändrad kapacitet Q_av', plotType='Elution-conductivity-vs-CV-combined-all')
for value in [1, 2, 3, 6]: par(Q_av=value); simu(4.0*V/VFR)

# Linje för 10% UV    
ax1.plot([0,4], [0.1,0.1],'k--')

# Återställ parametrar
par(Q_av=6.0, A_in=1.0)
par(start_adsorption = 1.0*V, stop_adsorption = 1.5*V)
par(start_desorption = 2.5*V, stationary_desorption = 5.5*V, stop_desorption = 7.5*V)
par(start_pooling = 3.7*V, stop_pooling = 7.0*V)
```

**Kommentar** Vi set att då kapaciteten Qav ökar så kommer genombrottet efter större mängd vätska som pumpats igenom. Notera att salt-koncentrationen går upp något under indbindningen men sedan går tillbaka till noll då bunder salt ersatts med produkt.

## 6 Variera gradientens lutning

```python
# Plot och simulering och gör A ännnu svagare för inbindning dvs minska k3
newplot(title='Inverkan av variation k3 - inbidningens styrka',\
        plotType='Elution-conductivity-vs-CV-combined-all')
for value in [5.5, 5.0, 4.5, 4.0]: 
    par(stationary_desorption=value*V); 
    simu(8.0*V/VFR)
    
# Restore default value
par(stationary_desorption=5.5*V)
```

**Kommentar** Med olika lutning på gradienten får vi olika grad av separation mellan topparna för $P$ resepektiva $A$. Nedan samma simulering med med fokus på UV. Här poolas med fast givna gränser utfrån pumpad volym vätska.

```python
# Simulering och plot i annat diagram med UV
newplot(title='Jämförelse fast poolning (-) med UV-poolning (--)', \
        plotType='Elution-vs-CV-pooling')
for value in [5.5, 5.0, 4.5, 4.0]: 
    par(stationary_desorption=value*V); 
    simu(8.0*V/VFR)
    
# Restore default value
par(stationary_desorption=5.5*V)
```

## 7 Byte av gradient till trappsteg för salt-buffert

```python
# Plot och simulering och gör A ännnu svagare för inbindning dvs minska k3
newplot(title='Inverkan av variation k3 - inbidningens styrka',\
        plotType='Elution-conductivity-vs-CV-combined-all')

par(gradient=False)
par(start_desorption = 2.3*V, x_start_desorption = 0.10)
par(stationary_desorption = 4.7*V)
par(start_pooling = 4.7*V, stop_pooling = 7.5*V)
simu(8.0*V/VFR)

#print('quality 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
#print('efficency 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))

par(E_in_desorption_buffer=16)
simu(8.0*V/VFR)
par(E_in_desorption_buffer=8)

# Restore default value
par(stationary_desorption=5.5*V)
par(stop_desorption=7.5*V)
par(gradient=True)
par(E_in_desorption_buffer=8)
```

## 8 Inverkan av salt koncentration i provet

```python
# Default example compared with a situation with salt in the sample
newplot(title='Hela cykeln: ekvilibrering, tvätt, desorption, poolning', \
        plotType='Elution-conductivity-vs-CV-combined-all')
simu(8.0*V/VFR)
par(E_in=2.0); simu(8.0*V/VFR)
par(E_in=4.0); simu(8.0*V/VFR)
par(E_in=0)
```

**Kommentar** I detta exempel verkar det som att salt i ingående sample i första hand tar plats från antagonistens inbindning. Detta skulle ju innebära en fördel. Medan inverkan på produkten är ganska liten. Kan det här stämma med verkigheten?

## 9 UV-poolning

```python
# Plot och simulering - nu plot där UV och poolning kan ses och kvalité och effektivitet beräknas
newplot(title='Jämförelse fast poolning (-) med UV-poolning (--)', plotType='Elution-vs-CV-pooling')

simu(8.0*V/VFR)
print('quality 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))

par(start_uv = 0.15, stop_uv = 0.001)
simu(8.0*V/VFR)
print('quality 2 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 2 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))

# Restore default values 
par(start_uv = -1, stop_uv = -2)
```

**Kommentar** Här jämförs UV-styrd poolning (streckad) jämfört med fast poolning styrd av volyms-schema (heldragen). Vi ser att poolning som är UV-styrd öppnar senare, men stänger vid samma volym. 

## 10 Variera inkommande material - fast poolning

```python
# Plot och simulering 
newplot(title='Jämförelse A=1.0 (-) med A=1.2 (--)', plotType='Elution-vs-CV-pooling')

par(start_pooling = 4.2*V, stop_pooling = 7.0*V)

par(P_in=1.0, A_in=1.0)
simu(8.0*V/VFR)
print('quality 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))

par(P_in=1.2, A_in=1.0)
simu(8.0*V/VFR)
print('quality 2 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 2 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))

# Restore default values 
par(A_in=1.0, start_pooling = 3.7*V)
```

**Kommentar** Vi ser resutatet av fast volyms-baserat schema för poolning och inverkan på mängd produkt som poolas då ingående mamterial får högre koncentration. 

## 11 Variera inkommande material - UV-poolning

```python
# Plot och simulering 
newplot(title='Jämförelse A=1.0 (-) med A=1.2 (--)', plotType='Elution-vs-CV-pooling')

par(start_uv=0.14, stop_uv=0.01)

par(P_in=1.0, A_in=1.0)
simu(8.0*V/VFR)
print('quality 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))

par(P_in=1.2, A_in=1.0)
simu(8.0*V/VFR)
print('quality 2 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 2 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))

# Restore default values 
par(A_in=1.0, start_uv=-1, stop_uv=-2)
```

**Kommentar** Vi ser att med UV-styrning kan poolningen göras mer följsam till förändringar i ingående mateias koncentration. I det här exemplet ser vi att mer material $P$ kan skördas med bibehållen kvalité. Resultatet ligger i presenterade siffror ovanför diagrammet i jämförelse med körningen tidigare med fast volym-baserat doseringsschema.

## 12 Slutsatser

Simuering med den här mycket förenklade dynamiska modellen av en jonbytare körd vid konstant $pH$ ger kvalitative förväntade resultat. Hur väl modellen kan kalibreras mot verkliga data återstår att undersöka.

Vi har också ett exempel på hur eventuellt $pH$ skulle kunna påverka de båda topparna och detta exempel behöver säkert förbättras om möjligt med rimlig ansträngning. Viktigt att förstå om $pH$ bara påverkar bindningens styrka ellar om den också påverkar kinetiken för de båda topparna. Hittills har vi fokuserat på bara inbindingens styrka.

Då modellen hela tiden använder mobilfas-volymen som grund för reaktioner har vi här en grov förenkling. Approximationen är rimlig för proteinerna och deras koncentrationer. Salt-koncentration borde dock räknas på en större volym än mobil-fasen och kanske rentav hela kolonnens volym. Det skulle till exempel göra att tiden för den initial ekvlibreringen blir cirka tre gånger längre vilket förmodligen stämmer bättre med verkligheten. Inverkan på topparna är svårare att bedöma.

Modellen beskriver väsentligen konvektion (transport) längs kolonnen samt bidning/frisläppning till resin. Här finns inget om vidare interaktion mellan mobilfas och resin där diffusion är en viktig mekanism. Den skulle kunna yttra sig genom att påverka bredden på topparna vad vi förstår fårn litteraturen.
