The interaction withe the model has changed slightly so that linear flow rate LF = F/area is used. The column volume is now specified in terms of (cross-section) area times height. The molecular weights listed below are not used in the simulations. They just give typical values and the molecular weight for bound protein and antagonist protein is just arbitrary here.

Parameters of the model in general as well as time scale are all arbitrary and focus is on qualitative aspects of the model.

The height of the column is 20 cm which is a common size in the industry.

## Loading

The parameter notation and values are the same as in the referred report. However the flow rate is here denoted F while q in the report. The column is diveded in n=8 sections and set at compilation time. The values are arbitrarly chosen in the report and the focus is on qualitative aspects of the model.

The simplified model describe only the column in terms of volume and does not distinguish a high column with a small diameter from a lower with larger diameter.

The parameters k1, k2, k3, k4 and Q_av are given relative volume and with increased column volume a larger capacity is thus obtained.

```python
describe('column.n')
```

```python
# Loading of the column - try to reproduce Jonas figure 13.
newplot(title='Adsorption along the column during loading', plotType='Loading') 

# Column properties 
par(LFR=12.0)
par(k1=0.3, k2=0.05, k3=0.05, k4=0.3, Q_av=3.0)
par(height=20, diameter=0.714)

# Operation
par(P_in=1.0, A_in=1.0, E_in=8)
par(scale_volume=False, stop_adsorption=50, start_desorption=150, stop_desorption=450)
par(start_pooling=220, stop_pooling=450)

# Simulation
simu(100)
```

```python
# We just check that we had the same volume flow rate as Jonas
describe('F')
```

Looks like Figure 13 in Jonas report!

```python
# Impact of shorter time for loading and then less material
newplot(title='Loading', plotType='Loading-combined')
show()

# Simulation with changed parameter t2
par(stop_adsorption=30); simu(100)

# Reset changed parameter
par(stop_adsorption=50)
```

To the left the inlet loading over time. 
To the right upper concentration along the column at different times and in steady states finally
To the right lower concentrations along the colunn in steady state.

We see that a shorter time and then less material makes less of the column capacity used.

Note that the flow through the column is constant despite change from sample to just buffer 1, and shown in how the volume of the waste tank increase with time.

## Change of resin property I

Here parameter k4 is lowered by a factor 6. This means that protein P still binds more easily to the gel than protein A but protein A now binds more easily to the gel than before.

```python
# Loading of the column - try to reproduce Jonas figure 15.
newplot(title='Adsorption along the column during loading - k4 changed', plotType='Loading') 

# Column properties 
par(diameter=0.714, k1=0.3, k2=0.05, k3=0.05, k4=0.05, Q_av=3.0)

# Operation
par(P_in=1, A_in=1.0)
par(LFR=12.0, stop_adsorption=50)

# Simulation
simu(100)
```

This corresponds well to Jonas figure 15.

The main impact of lowering parameter k4 is that concentration of bound A doubles.
Also some P is not bound and therefore wasted.

```python
# Elution of the column - try to reproduce figure 16
newplot(title='Elution peaks - parameter k4 is changed', plotType='Elution')

# Operation
par(P_in=1, A_in=1.0, E_in=8)
par(LFR=12.0, stop_adsorption=50, start_desorption=150, stop_desorption=450)

# Simulation
simu(550)
ax2.set_ylim([0, 0.6])
plt.show()
```

Results correspond well to figure 16.

Now we see no separation between the two components P and A in the chromatogram.

## When does superposition of chromatograms hold?

```python
# Loading and elution of the column - try to reproduce Jonas figure 18.
newplot(title='Elution peaks - individual and combined - load 1.0 g/L each', plotType='Elution') 

# Column properties 
par(diameter=0.714, k1=0.3, k2=0.05, k3=0.05, k4=0.3, Q_av=3.0)

# Operation
par(LFR=12.0, stop_adsorption=50, start_desorption=150, stop_desorption=450)

# Simulation
par(P_in=1.0, A_in=1.0); simu(550)
par(P_in=1.0, A_in=0.0); simu(550)
par(P_in=0.0, A_in=1.0); simu(550)

# Adjust diagrams
ax1.set_ylim([0, 0.65])
ax2.set_ylim([0, 0.65])
plt.show()
```

We see that the sample with both P and A gives a chromatogram (solid line) that is different from what we would get from adding chromatograms for the individual runs (dashed and dottet line). 

The main reason is that the high load of the column makes the response to A very different when run individually.

In this case with high load on the column superposition does not hold.

```python
# Now contiune with a much lower load - try to reproduce Jonas figure 18.
newplot(title='Elution peaks - individual and combined - load 0.1 g/L each', plotType='Elution') 

# Column properties 
par(diameter=0.714, k1=0.3, k2=0.05, k3=0.05, k4=0.3, Q_av=3.0)

# Operation
par(LFR=12.0, stop_adsorption=50, start_desorption=150, stop_desorption=450)

# Simulation
par(P_in=0.1, A_in=0.1); simu(550)
par(P_in=0.1, A_in=0.0); simu(550)
par(P_in=0.0, A_in=0.1); simu(550)

# Adjust diagrams
#ax1.set_ylim([0, 0.65])
#ax2.set_ylim([0, 0.65])
plt.show()
```

We now see with this lower load that chromatogram can be seen as a superposition of the chromatograms for the individual components. Result the same as figure 18.

## Pooling
We test pooling of product based on given times, i.e. without feedback from UV.
The material goes to tank_waste except during time for pooling when it goes to tank_harvest.
Start with the settings of the first example.

The simulations below are not found in Jonas report but natural to investigate a bit.

```python
# Loading and elution as for the very first example and now with pooling at the outlet
newplot(title='Elution and pooling', plotType='Elution-pooling') 

# Column properties 
par(diameter=0.714, k1=0.3, k2=0.05, k3=0.05, k4=0.3, Q_av=3.0)

# Operation
par(P_in=1, A_in=1.0)
par(LFR=12.0, stop_adsorption=50, start_desorption=150, stop_desorption=450)
par(start_pooling=(150+80), stop_pooling=(150+300))

# Simulation
simu(550)

plt.show()
```

```python
# And how material is accumulated in harvest and waste tanks, respectively
newplot(title='Pooling to tank harvest and waste', plotType='Pooling'); show()
```

```python
# Quality of the harvested product
print(model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))

# Efficiency of collecting the protein of interest calculated below
print(model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))
```

We have a quality of 98% and an efficiency of about 95 percent and looks good!

Note, that a small shift in the prescribed pooling intervall will change both quality and the efficiency as seen below.

```python
# Loading and elution as above but now compared with a small shift of pooling
newplot(title='Elution and pooling', plotType='Pooling') 

# Column properties 
par(diameter=0.714, k1=0.3, k2=0.05, k3=0.05, k4=0.3, Q_av=3.0)

# Operation
par(P_in=1, A_in=1.0)
par(LFR=12.0, stop_adsorption=50, start_desorption=150, stop_desorption=450)
par(start_pooling=(150+80), stop_pooling=(150+300))

# Simulations
simu(550)

par(start_pooling=(150+90)); simu(550)
```

```python
# Quality of the harvested product
print('quality =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))

# Efficiency of collecting the protein of interest calculated below
print('efficency =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))
```

As we change the pooling to start later, the product quality will increase but the efficiency of the separation process goes down.

## Pooling and impact of changes in the column resin on results

```python
# Loading, elution and fixed time for pooling and impact of changed resin 
newplot(title='Elution and pooling', plotType='Elution-pooling') 

# Column properties 
par(diameter=0.714, k1=0.3, k2=0.05, k3=0.05, k4=0.3, Q_av=3.0)

# Operation
par(P_in=1, A_in=1.0)
par(LFR=12.0, stop_adsorption=50, start_desorption=150, stop_desorption=450)
par(start_pooling=(150+80), stop_pooling=(150+300))

# Simulations
simu(550)
print('quality 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))

par(k4=0.2); simu(550)
print('quality 2 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 2 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))
```

We see that a change in the resin property affect quality(and also efficency, not shown). The best would be to adapt the pooling limits to the chromatogram, i.e. use feedback control of measured UV-signal.

```python
# Loading, elution and fixed time for pooling and impact of changed resin 
newplot(title='Elution and pooling', plotType='Elution-pooling') 

# Column properties - and now k4 remains changed
par(diameter=0.714, k1=0.3, k2=0.05, k3=0.05, k4=0.2, Q_av=3.0)

# Operation
par(P_in=1, A_in=1.0)
par(LFR=12.0, stop_adsorption=50, start_desorption=150, stop_desorption=450)
par(start_pooling=(150+80), stop_pooling=(150+300))

# Simulations - compare t_start_pool fixed and adjusted 
simu(550)
print('quality 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 1 =',model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))

par(t_start_pooling=(150+92)/60)
simu(550)
print('quality 1 =', model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_harvest.m[2]')))
print('efficency 2 =', model.get('tank_harvest.m[1]')/(model.get('tank_harvest.m[1]') + model.get('tank_waste.m[1]')))
```

So by start pooling after the chromatogram has passed the local minima and use that as a control principle, we in this example decrease the impact of resin degradation on product quality. The price is lowered efficency, though.

## Decrease impact of resin degradation using feedback control

The idea of start pooling after the chromatogram has passed the local minima can be automated and tested in this section. The idea is to simply measure the UV on-line and open and close the valve for pooling. The key problem is to translate the idea of identify the local mininima in a robust way. In this example here is no measurement noise and we implement a rudiementary controller and just reproduce the results from previous section and check also impact of larger resin degradation. 

In practice, a routine similar like this one is often used. The actual guarantee of preserving quality this way must be investigated experimentally.
