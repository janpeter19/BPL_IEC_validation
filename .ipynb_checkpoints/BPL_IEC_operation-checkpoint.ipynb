## Upprepa inledande cell 3 ovan och inför mer korrekt pH-hantering

Vi gör följande om-parametrisering från $k_1$,  $k_2$ till $K_P$ och $T_P$ då de senare har en tydligare tolkning.
Parametrarna för antagonisten görs på liknande sätt.

$$
\eqalign{
K_P &= k_1/k_2 \cr
T_P &= 1/k_1
}
$$

Förändring i skillanden $pI_P$ - $pH_{resin}$ påverkar $K_P$ proportionellt. Men notera att $K_P$>0 då både $k_1$>0 och $k_2$>0 är det ända rimliga kemiskt sett.

Tidskonstanten $T_P$ påverkas inte av detta, men beror av flera andra faktorer.

Nedan definierar iv ett par hjälpfuntioner för hur $K_P$ och $K_A$ påverkas. Jag har valt värden sår att vi får tillbaka Jonas värden på k1, k2, k3, k4 med nominella värden på $pI_P$, $pI_A$ och $pH_{resin}.$


```python
# Definiera hur skillnande pI_P och pI_A med pH_resin påverkar K_P och K_A

def KP_pH_sensitivity(pI_P=8.0, pH_resin=7.0):
    K_P_nom = 0.0
    delta_pH = 6.0
    return K_P_nom + delta_pH*(pI_P-pH_resin)

def KA_pH_sensitivity(pI_A=7.1667, pH_resin=7.0):
    K_A_nom = 0.0
    delta_pH = 1.0
    return K_A_nom + delta_pH*(pI_A-pH_resin)
```

```python
TP = 3.33
TA = 20.0

k1_value = 1/TP
k2_value = 1/(TP*KP_pH_sensitivity())
k3_value = 1/TA
k4_value = 1/(TA*KA_pH_sensitivity())
```

```python
print(k1_value, k2_value)
print(k3_value, k4_value)
```

Nu har vi rudimentärt testat att räkningnarna fungerar.

Dags att använda dem!

```python
# Körning med rimliga parametrar för körning med bibehållna parametrar för gelen k1..k4 men Q_av ökad

# Beräkning av V och VFR utifrån givet h, d, lfr
from numpy import pi
h = 20.0
d = 1.27
a = pi*(d/2)**2
V = h*a
#print('V =', np.round(V,1), '[mL]')
lfr = 48
VFR = a*lfr/60
#print('VFR =', np.round(VFR,1), '[mL/min]')

# Bekväma parametrar för omstlagspunkter givet kolonnvolym V beräknat ovan
CV_ekv = 1.0
CV_ads = 0.5
CV_wash = 1.0
CV_desorb = 3.0
CV_start_pool = 1.2
CV_stop_pool = 4.5
CV_wash2 = 2.5

# Exempel på parametrar
TP = 3.33
TA = 20.0

# Parametrar
par(k1 = 1/TP)
par(k2 = 1/(TP*KP_pH_sensitivity(pI_P=7.00001, pH_resin=7.0)))
par(k3 = 1/TA)
par(k4 = 1/(TA*KA_pH_sensitivity(pI_A=7.16667, pH_resin=7.0)))
par(Q_av = 6.0)
init(E_0 = 50)
par(P_in = 1.0)
par(A_in = 1.0)
par(E_in_desorption_buffer = 8.0)
par(LFR=lfr)
par(height = h)
par(diameter = d)
par(scale_volume=True, start_adsorption=CV_ekv*V, stop_adsorption=(CV_ekv+CV_ads)*V)
par(start_desorption=(CV_ekv+CV_ads+CV_wash)*V, stationary_desorption=(CV_ekv+CV_ads+CV_wash+CV_desorb)*V)
par(stop_desorption=7.5*V)
par(start_pooling=(CV_ekv+CV_ads+CV_wash+CV_start_pool)*V, stop_pooling=(CV_ekv+CV_ads+CV_wash+CV_stop_pool)*V)

# Plot och simulering
newplot(title='Adsorption längs kolonnen - olika pI_P', plotType='Elution-conductivity-vs-CV-combined-all')

par(k2 = 1/(TP*KP_pH_sensitivity(pI_P=7.2, pH_resin=7.0))); 
simu((CV_ekv+CV_ads+CV_wash+CV_desorb+CV_wash2)*V/VFR)

par(k2 = 1/(TP*KP_pH_sensitivity(pI_P=7.6, pH_resin=7.0))); 
simu((CV_ekv+CV_ads+CV_wash+CV_desorb+CV_wash2)*V/VFR)

par(k2 = 1/(TP*KP_pH_sensitivity(pI_P=8.0, pH_resin=7.0)))
simu((CV_ekv+CV_ads+CV_wash+CV_desorb+CV_wash2)*V/VFR)
```

```python
# När pI_P kommer mycket nära pH_resin får vi minimal inbindning av P och påverkar inbindning A något

# Plot och simulering
newplot(title='Adsorption längs kolonnen - olika pI_P nära pH_resin', 
        plotType='Elution-conductivity-vs-CV-combined-all')

par(k2 = 1/(TP*KP_pH_sensitivity(pI_P=7.0001, pH_resin=7.0))); 
simu((CV_ekv+CV_ads+CV_wash+CV_desorb+CV_wash2)*V/VFR)

```
